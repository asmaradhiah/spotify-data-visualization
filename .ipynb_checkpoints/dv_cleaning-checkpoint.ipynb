```python
# Remove rows with missing streams
df = df.dropna(subset=['streams'])

# Convert streams to integer
df['streams'] = df['streams'].astype('int64')

# Remove duplicates
df = df.drop_duplicates()
```

```python
df.to_csv(
    'dataset/Most_Streamed_Spotify_Songs_2023_Cleaned.csv',
    index=False
)
```

```python
#Exploratory Data Analysis Phase (EDA)
```

```python
df.describe() #to know basic stats
```

```python
#check data distribution
import matplotlib.pyplot as plt

plt.hist(df['streams'], bins=30)
plt.title('Distribution of Streams')
plt.xlabel('Streams')
plt.ylabel('Frequency')
plt.show()
```

```python
#Top 10 Streamed Songs
top10 = df.nlargest(10, 'streams')

top10[['track_name', 'artist(s)_name', 'streams']]

top10.plot(
    x='track_name',
    y='streams',
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Most Streamed Songs')
plt.show()
```

```python
#Top Artist by Streams
artist_streams = (
    df.groupby('artist(s)_name')['streams']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
artist_streams
top10.plot(
    x='artist(s)_name',
    y='streams',
    kind='barh',
    figsize=(10,5)
)

plt.title('Top 10 Most Streamed Artists')
plt.show()
```

```python
#Release Trends: Tahun mana paling banyak lagu dalam dataset?
songs_by_year = df['released_year'].value_counts().sort_index()

songs_by_year.plot(kind='line')
plt.title('Songs Released by Year')
plt.show()
```

```python
#Correlation Analysis: Untuk tengok faktor yang berkaitan dengan streams.

import seaborn as sns

plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

```python
#We can see that the highest correlation is in spotify playlist with streams 
sns.scatterplot(
    data=df,
    x='in_spotify_playlists',
    y='streams'
)

plt.show()
```

```python
#Audio features analysis
features = [
    'danceability_%',
    'energy_%',
    'valence_%',
    'acousticness_%'
]

df[features].mean()

df[features].mean().plot(kind='bar')
plt.title('Average Audio Features')
plt.show()
```

```python
#Distribuiton of streams, bpm and danceability
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Streams
sns.histplot(df['streams'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Streams')

# BPM
sns.histplot(df['bpm'], bins=20, kde=True, color='orange', ax=axes[1])
axes[1].set_title('Distribution of BPM')

# Danceability
sns.histplot(df['danceability_%'], bins=20, kde=True, color='green', ax=axes[2])
axes[2].set_title('Distribution of Danceability (%)')

plt.tight_layout()
plt.show()
```
